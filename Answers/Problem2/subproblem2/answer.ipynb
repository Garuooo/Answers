```python
import pandas as pd
```

```python
table_one = pd.read_csv("table1.csv",sep="\t")
```

```python
table_one
```

```python
table_two = pd.read_csv("table2.csv",sep="\t")
```

```python
table_two
```

```python
table_two["key1"]=table_two["key1"].str.upper()
table_two["key2"]=table_two["key2"].str.upper()
table_two["R"]=table_two["R"].str.upper()
table_two["S"]=table_two["S"].str.upper()
```

```python
inner_merged_data = pd.merge(table_one,table_two,on=["key1","key2"],how='inner')
```

```python
inner_merged_data
```

```python
left_merged_data = pd.merge(table_one,table_two,on=["key1","key2"],how='left')
```

```python
left_merged_data
```

```python
left_merged_data.fillna(0,axis=0,inplace=True)
```

```python
left_merged_data
```

```python
right_merged_data = pd.merge(table_one,table_two,on=["key1","key2"],how='right')
```

```python
right_merged_data
```

```python
right_merged_data.fillna(0,axis=0,inplace=True)
```

```python
right_merged_data
```

```python
outer_merged_data = pd.merge(table_one,table_two,on=["key1","key2"],how="left")
```

```python
outer_merged_data
```

```python
outer_merged_data.fillna(0,axis=0,inplace=True)
```

```python
outer_merged_data
```
